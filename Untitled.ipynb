```python
import numpy as np
from scipy.stats import gamma
import matplotlib.pyplot as plt
import math
import scipy.stats as stats
```

```python
mean = 5
variance = 10
loc = -1

beta = (mean/variance)**(1./3.)
alpha = mean / beta
x = np.linspace(gamma.ppf(0.01, alpha), gamma.ppf(0.99, alpha), 100)
y = gamma.pdf(x, a=alpha, scale=1/beta, loc=loc)
fig, ax = plt.subplots(1, 1)
ax.plot(x, y,'r-', label='gamma pdf')
```

```python
# Dauer der Infektiösität. Laut RKI zwischen 1-2 Tage vor und bis 8 Tage nach Symptombeginn
# Daraus ergibt sich ein Wert für Mean = 9

mu = 9
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```
